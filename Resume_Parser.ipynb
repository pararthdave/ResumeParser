```python
import spacy
from spacy.matcher import Matcher
import pdftotext
import pandas as pd
import os
```

```python
with open("/content/drive/MyDrive/MyDocuments/ResumeParser/data/data/data/ADVOCATE/20544228.pdf", "rb") as f:
    pdf = pdftotext.PDF(f)
```

```python
finalpg=[]
for page in pdf:
    pg=page.strip()
    finalpg.append(pg)
print(finalpg[0])
```

Now that the text has been imported into the environment. Let us try to extract the needed information from the text using Spacy library.

First, let us look at the tags assigned by Spacy to each word in the document.

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(finalpg[0])


for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```
